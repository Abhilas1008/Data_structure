```python
#1: Discuss string slicing and provide examples.
''' 
Slicing :
Ans: You can return range of characters by using the slice syntex.
Specify the start index and the end index, separated by a colon, to return a part of the string.
'''# Example "
#  Get the characters from position 2 to position 5(not included):
b = "hello abhi"
print(b[2:5])


'''Slicing from the start:
by leaving out the start index , the range will start at the first character:
'''# Example:
# get the characters from the start to position 5(not included):
b = "hello abhi"
print(b[:5])


'''Slicing to the end:
by leaving out the end index , the range will go to the end:
'''# Example:
# get the characters from position 2, and all the way to the end:
b = "hello abhi"
print(b[2:])


'''negative indexing:
use nrgative index to start the slice from the end of the string
'''# Example:
# get the characters :
# from "h" in "abhi"(position -2) to, but not included : "h" in "abhi" (position -2)
b = "hello abhi"
print(b[-5:-2])

```

```python
#6: Discuss the use cases of tuples and sets in python programming.
'''
Ans:
#USE CASES OF SETS:

#1
How to Use Sets to Remove Duplicate Items in Other Data Structures
We can use sets to get rid of duplicate items in other data structures like lists and tuples.

This is useful when you are dealing with a very large data set that requires only the individual unit of items returned 
instead of the number of occurrences of the items '''
#example
lis = [3,4,5,6,3]
lis = set(lis)
print(lis)
'''In the code above, we create a list of numbers called lis which had multiple occurrences of some of the numbers: [3,4,5,6,3].

Using the set() method, we converted the list lis to a set

When the new set was printed out, all the duplicates of the numbers were removed '''

'''
2#
Perform Mathematical Operations Using Sets in Python
Sets in Python are similar to sets in mathematics, and we can get various results based on the relationship that exists 
between sets.
union, intersection, difference and symmetric difference.'''
#union
'''Get the Union of Sets in Python
The union of two sets is the set of all the individual items that exist in both sets. In the union, duplicates are ignored.'''
firstSet = {2, 3, 4, 5}
secondSet = {1, 3, 5, 7}
print(firstSet | secondSet)
'''In the example we have two sets – firstSet = {2, 3, 4, 5} and secondSet = {1, 3, 5, 7}.
Using the vertical bar operator (|)we get the union of the two sets: firstSet | secondSet.
output {1, 2, 3, 4, 5, 7}. You can see that the two sets have been combined to form one set without any duplicates.'''
#intersection
'''The intersection of two sets is the set of items that are common to both sets. 
it is the set of items that appears in both firstSet and secondSet.'''
firstSet = {2, 3, 4, 5}
secondSet = {1, 3, 5, 7}
print(firstSet & secondSet)
'''we're using the & operator to get the intersection of firstSet and secondSet: firstSet & secondSet.
We got {3, 5} returned because 3 and 5 appear in both sets.'''

#Difference
'''The difference between two sets is the set of all the items that exist in one set but not the other.'''
firstSet = {2, 3, 4, 5}
secondSet = {1, 3, 5, 7}
print(firstSet - secondSet)
'''In the example, we are getting a set of the items that exist in firstSet but not in secondSet.
We used the - operator to achieve this: firstSet - secondSet.
output from the operation is 2 and 4.'''

#Symmetric Difference 
'''The symmetric difference of two sets is the set of all items that exist in either of the two sets but not in both.
The symmetric difference is the result of items that exist in each set but not in both.'''
firstSet = {2, 3, 4, 5}
secondSet = {1, 3, 5, 7}
print(firstSet ^ secondSet)

'''We used the ^ operator to get the symmetric difference of two sets: firstSet ^ secondSet.
output was 1, 2, 4, 7.  Each of these items do not appear in both sets.'''

'''#USE CASES OF TUPLE:
#Tuples are great, because of mostly two things:
Tuples are considered heterogenous data structures, meaning there's no implicit expectation that the contents are all the same
type. Furthermore, they signal meaningful positioning; for instance the row of a table can be treated as a tuple, because one 
column could be holding names and another a person's age. Swapping them around in a row could end up breaking things.

#Tuples are immutable, and also hashable, meaning they can be used as dictionary keys. The immutability part also enforces that
no part of your code can accidentally add or remove elements from it. 
And if the contents are also immutable, you can rest assured that it cannot change at all unless you assign a different tuple 
to the same name.'''
```

```python
#7 : Describe how to add, modify , delete item in dictionary with examples.
'''
Ans:
'''
#adding
devBio = {
  "name": "Ihechikara",
  "age": 120,
  "language": "JavaScript"
}

devBio["role"] = "Developer"

print(devBio)
#modify
devBio = {
  "name": "Ihechikara",
  "age": 120,
  "language": "JavaScript"
}

devBio.update({"role": "Developer"})

print(devBio)



#delete
'''Using pop() method on dictionary:
The pop() function may be used to delete any key from Python. A dictionary's pop() function is used to remove a key: value pair.
The method outputs the string that we can supply as an additional parameter and returns the removed pair if it doesn't exist
(an exception is thrown otherwise).
'''
langDict = {"python" : 24, "c++" : 22, "javaScript" : 25, "java" : 20, "r" : 21 } 
print(langDict) 
langDict.pop("python")
print(langDict) 

'''Using del method:
Using the del function is another way to remove a key from a dictionary. To remove a particular key from the dictionary,
use the del keyword and the dict[key] notation. In the event of a false value, del does not allow any further value to be
returned. Instead, it will produce an error.'''

langDict = {"python" : 24, "c++" : 22, "javaScript" : 25, "java" : 20, "r" : 21 } 
print("The dictionary is :", langDict)  
del langDict["python"] 
print("Deleting Key ", langDict)

'''Using iterator/comprehension:'''
langDict = {"python" : 24, "c++" : 22, "javaScript" : 25, "java" : 20, "r" : 21 } 
print("The dictionary is :", langDict) 
erase = input("Enter the key to be deleted ") 
newDict = { key:val for key, val in langDict.items() if key != erase} 
print("Deleting Key ", newDict)
```
